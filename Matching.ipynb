```python
import pandas as pd
import re
from nltk.corpus import stopwords

# Téléchargez les stopwords pour la langue arabe si ce n'est pas déjà fait
import nltk
nltk.download('stopwords')

# Fonction pour nettoyer le texte
def nettoyer_texte(texte):
    # Supprimer les hashtags
    texte = re.sub(r'#', '', texte)
    
    # Supprimer les emojis (vous pouvez ajuster cette expression régulière en fonction de vos besoins)
    texte = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F780-\U0001F7FF\U0001F800-\U0001F8FF\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U000024C2-\U0001F251]', '', texte)
    
    # Supprimer les parenthèses, guillemets doubles et simples
    texte = re.sub(r'[()\[\]{}"\'‘’“”]', '', texte)
    
    # Supprimer les nombres
    texte = re.sub(r'\d+', '', texte)
    
    # Supprimer les caractères spécifiques
    texte = re.sub(r'[|\\><]', '', texte)
    
    # Supprimer les stopwords
    stop_words_arabe = set(stopwords.words('arabic'))
    mots = texte.split()
    mots = [mot for mot in mots if mot.lower() not in stop_words_arabe]
    
    # Rejoindre les mots nettoyés en une chaîne
    texte_nettoye = ' '.join(mots)
    
    return texte_nettoye
```

```python

# Appliquer la fonction de nettoyage à la colonne contenant des mots en arabe
df1['Titre'] = df1['Titre'].apply(nettoyer_texte)
df1['Date_exacte'] = pd.to_datetime(df1['Date_exacte'])
df1['Date_exacte'] = df1['Date_exacte'].dt.date
df1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def calculate_similarity(text1, text2):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([text1, text2])
    similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
    similarity_text = similarity_matrix[0, 1]
    return similarity_text
```

```python
def calculate_date_similarity(date1, date2):
    return date1 == date2
```

```python
similarity_results = []

for index_df1, row_df1 in df1.iterrows():
    for index_df2, row_df2 in df2.iterrows():
        similarity_text = calculate_similarity(row_df1["Titre"], row_df2["title"])
        similarity_date = calculate_date_similarity(row_df1['Date_exacte'], row_df2['video_published'])
        
        if similarity_text >= 0.6 and similarity_date:
            similarity_results.append({
                'date_df1': row_df1['Date_exacte'],
                'date_df2': row_df2['video_published'],
                'titre_df1': row_df1['Titre'],
                'titre_df2': row_df2['title'],
                'similarity_text': similarity_text,
                'similarity_date': similarity_date
            })

similarity_df = pd.DataFrame(similarity_results)
```
